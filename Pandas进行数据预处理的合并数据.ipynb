```python
#主键合并
order = pd.read_csv('./meal_order_info.csv',encoding='gbk')
order.head()
```

```python
order['info_id'].dtype
```

```python
data['order_id'].dtype
```

```python
#强制转换
order['info_id'] = order['info_id'].astype(str)
```

```python
#主键合并数据
pd.merge(order,data,left_on='info_id',right_on='order_id').shape#数据合并，两个表的合并，通过两个表的主键进行合并数据，通过.shape方法查看行列
```

```python
data.shape#行数取最大值，而列没有相同的列名称，则列相加
```

```python
#order#修改两个表的主键名，需保持一致，修改之前
order.rename(columns={'info_id':'order_id'},inplace=True)#对order里面的info_id修改为order_id，但因inplace为True，所以order里面数据产生影响
#order#修改之后
data.join(order,rsuffix='l')#这里的on参数不添加，且rsuffix为右边堆叠，将两个表合并
#与merge方法结果一样，但是需要命名同一个主键名
```

## 重叠合并数据

### 一张表数据完整，一张表数据缺失的时候，进行重叠数据合并

```python
#采用combine_first方法，使用方法：pandas.DataFrame.combine_first(other)
import numpy as np
import pandas as pd
dis1 = {'id':list(range(1,10)),
       'cpu':['i7','i5',np.nan,'i7','i7','i5',np.nan,np.nan,'i5']}
a = pd.DataFrame(dis1)#转成数据框
dis2 = {'id':list(range(1,10)),
       'cpu':['i7','i5','i5',np.nan,'i7','i5','i5',np.nan,'i5']}
b = pd.DataFrame(dis2)
```

```python
dis1
```

```python
#进行拼接合并
a.combine_first(b)#消除缺失值
```
